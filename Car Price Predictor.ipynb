# Cleaning Data

```python
car=car[car['year'].str.isnumeric()]
# keeping all the values where the values inside the string is numeric
```

```python
car
```

```python
car['year'].unique()
```

```python
car['year']=car['year'].astype(int)
# converting the string values to numeric
```

```python
car=car[car['Price']!='Ask For Price']
# Removing the Rows where the Price column has 'Ask For Price'
```

```python
car['Price'].unique()
```

```python
car['Price']=car['Price'].str.replace(',','').astype(int)
# Converting the Price values having comma to empty string to maintain consistency
```

```python
car['Price'].unique()
```

```python
car['kms_driven']=car['kms_driven'].str.split().str.get(0).str.replace(',','')
# removing 'kms' from the end of the values
```

```python
car['kms_driven'].unique()
```

```python
car=car[car['kms_driven'].str.isnumeric()]
# conveting to numeric values
```

```python
car['kms_driven']=car['kms_driven'].astype(int)
```

```python
car['kms_driven'].unique()
```

```python
car=car[~car['fuel_type'].isna()]
# keeping the values where fuel_type is not nan
```

```python
car['fuel_type'].unique()
```

```python
car.info()
```

```python
car['name']=car['name'].str.split().str.slice(start=0,stop=3).str.join(' ')
# only using the first three names of the car name 
```

```python
car=car.reset_index(drop=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import make_pipeline
```

```python
x=car.drop(columns="Price")
y=car['Price']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
ohe=OneHotEncoder()
ohe.fit(x[['name','company','fuel_type']])
```

```python
ohe.categories_
```

```python
column_trans=ColumnTransformer(transformers=[
    ('ohe',OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),
],remainder='passthrough')
```

```python
lr=LinearRegression()
```

```python
pipe=make_pipeline(column_trans,lr)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred=pipe.predict(x_test)
```

```python
y_pred
```

```python
r2_score(y_test,y_pred)
```

```python
scores=[]
for i in range(10000):
    X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(X_train,y_train)
    y_pred=pipe.predict(X_test)
    scores.append(r2_score(y_test,y_pred))
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```
